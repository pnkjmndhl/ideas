```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#let x,y are the coordinates of stores
points = np.array([[5, 3], [10, 15], [15, 12], [24, 10], [30, 45], [85, 70], [71, 80], [60, 78], [55, 52],[80, 91]])

xs = points[:,0] # Selects all xs from the array
ys = points[:,1]  # Selects all ys from the array

plt.title("10 Stores Coordinates")
plt.scatter(x=xs, y=ys)
```

```python
import scipy.cluster.hierarchy as sch

dndgram = sch.dendrogram(sch.linkage(points, method = "ward"))
plt.title ("Dendogram")
plt.xlabel("Coordinates")
plt.ylabel("Euclidean Distances")
#plt.xlabels(points)
plt.show()

#check for the largest distance vertically without crossing any horizontal line
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

cluster = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
hcl = cluster.fit_predict(points)
ss = silhouette_score(points, hcl)
print(ss)
```

```python
#plotting the clusters
plt.scatter(points[cl ==0, 0], points[cl ==0, 1], c = 'red', label = 'cluster1')
plt.scatter(points[cl ==1, 0], points[cl ==1, 1], c = 'green', label = 'cluster2')
plt.scatter(points[cl ==2, 0], points[cl ==2, 1], c = 'blue', label = 'cluster3')
```
