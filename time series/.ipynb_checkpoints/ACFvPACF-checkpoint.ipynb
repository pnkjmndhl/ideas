```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
register_matplotlib_converters()
```

```python
temp_df = pd.read_csv('../data/daily-min-temperatures.csv')

#view dataset sample
temp_df.head()
```

```python
#convert date column to datetime type
temp_df['Date'] = pd.to_datetime(temp_df.Date)
#set date as index
temp_df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(10,4))
plt.plot(temp_df.Temp)
plt.title('Temperature over time', fontsize=20)
plt.ylabel('Temperature', fontsize=16)
for year in range(1981,1991):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

#### ACF

```python
acf_plot = plot_acf(temp_df.Temp, lags = 1000)
```

#### Based on rising ACF, we are likely dealing with an autoregressive process

#### PACF

```python
pacf_plot = plot_pacf(temp_df.Temp)
```

### Stock Data

```python
import yfinance as yf

#define the ticker symbol
tickerSymbol = 'SPY'
#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)
#get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2015-1-1', end='2022-8-20')
tickerDf = tickerDf[['Close']]
#see your data
tickerDf.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(tickerDf.Close)
plt.title('Stock Price over Time (%s)'%tickerSymbol, fontsize=20)
plt.ylabel('Price', fontsize=16)
for year in range(2015,2023):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

#### Stationarity: take first difference of the series

```python
#take first difference
first_diffs = tickerDf.Close.values[1:] - tickerDf.Close.values[:-1]
first_diffs = np.concatenate([first_diffs, [0]])
#set first difference as variable in dataframe
tickerDf['FirstDifference'] = first_diffs
tickerDf.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(tickerDf.FirstDifference)
plt.title('First Difference over Time (%s)'%tickerSymbol, fontsize=20)
plt.ylabel('Price Difference', fontsize=16)
for year in range(2015,2021):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

#### ACF

```python
acf_plot = plot_acf(tickerDf.FirstDifference)
```

#### ACF isn't that informative

#### PACFACF isn't that informative

```python
pacf_plot = plot_pacf(tickerDf.FirstDifference)
```
